```python
file = open('Gaurav.txt','r')
print(file.read())
```

```python
with open('Gaurav.txt', 'r') as file:
    print(file.read())
```

```python
#reading by charecters - " " is a  charecter as well.

with open('Gaurav.txt', 'r') as file:
    print(file.read(20))
```

```python
#readline

with open('Gaurav.txt', 'r') as file:
    print(file.readline())

file.close()
```

```python
#to read the file using readline

file = open('Gaurav.txt','r')

line = file.readline()

while line:
    print(line)
    line= file.readline()

file.close()

```

```python
#returns the lines as list of strings

with open('Gaurav.txt', 'r') as file:
    data = file.readlines()
    print(data)
    for i in data:
        print(i.split())

file.close()
```
