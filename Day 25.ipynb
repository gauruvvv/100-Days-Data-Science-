```python
#numpy slicing
'''Difference  between slicing a list and slicing an array

while slicing an array, if you say

b = a[1:5]
b makes a copy of a, so if you change anything in b, it 
will also change in a.

unlike list slicing

slicing and indexing methods still remain the same though
'''
  
```

```python
a = np.arange(100)
```

```python
b=a[3:10]
print(b)
```

```python
b[3] = 5

print(b)
print(a)

'''3rd index has changed in both a and b'''
```

```python
b[0] = -1200

print(a)
print(b)
```

```python
A = np.arange(1000)
print(A)
```

```python
#if you want to make an independent array b

b = A[3:10].copy()
```

```python
#A remains the same

b[3] = -1200
print(b)
print(A)
```
